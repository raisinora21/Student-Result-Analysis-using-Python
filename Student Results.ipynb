```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
print (df.head())
```

```python
#mandatory steps
df.describe()
```

```python
df.info()
```

```python
#gender distribution
custom_palette = ["blue", "orange"]
plt.figure(figsize= (5,5))

ax = sns.countplot(data=df, x="Gender", hue="Gender", palette=custom_palette)

ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")

plt.show()
```

```python
#from the above chart we have analysed that 
#the number of females in the data is more than number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore": 'mean', "ReadingScore": 'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize= (4,4))
sns.heatmap(gb, annot= True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

```python
#from the above chart we have concluded that the education of the parents have a good impact on their scores
```

```python
gb1 =  df.groupby("ParentMaritalStatus").agg({"MathScore": 'mean', "ReadingScore": 'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize= (4,4))
sns.heatmap(gb1, annot= True)
plt.title("Relationship between Parent's Marital Status and Student's Score")
plt.show()
```

```python
#from the above chart we can see that there is no/neglible impact on the 
#student's score due to their parent's marittal status
```
